```python
import numpy as np
```

```python
def print_mat(data, thresh=0.8):
    for i in range(32):
        line = ''
        for k in range(32):
            if float(data[i,k]) > thresh:
                line += '1'
            else:
                line += '0'
        print(line)
```

```python
def compress_image(num_sv = 3, thresh = 0.8):
    output = []
    for line in open('data.txt').readlines():
        new_row = []
        for i in range(32):
            new_row.append(int(line[i]))
        output.append(new_row)
    
    mat = np.mat(output)
    print('original matrix')
    print_mat(mat, thresh)
    
    u, sigma, vt = np.linalg.svd(mat)
    sigma_reconstructed = np.mat(np.zeros((num_sv, num_sv)))
    for k in range(num_sv):
        sigma_reconstructed[k, k] = sigma[k]
    reconstructed_matrix = u[:, :num_sv] * sigma_reconstructed * vt[:num_sv, :]
    print()
    print('reconstructed matrix using {} singular values'.format(num_sv))
    print_mat(reconstructed_matrix, thresh)
```

```python
compress_image(2)
```
